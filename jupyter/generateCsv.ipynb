```python
import numpy as np
def random_dates(start, end, n, freq, seed=None):
    if seed is not None:
        np.random.seed(seed)

    dr = pd.date_range(start, end, freq=freq)
    return pd.to_datetime(np.sort(np.random.choice(dr, n, replace=False)))

random_dates('2024-04-01', '2024-04-30', 91, 's', seed=None)
```

# Using this to quickly generate 1K game_sales records.

```python
import random
import pandas as pd
df = pd.read_csv('../src/main/resources/100GameTitle.csv')


arTime = random_dates('2024-04-01', '2024-04-30', 100, 's', seed=None)
df["date_of_sale"]=df.apply(lambda row: arTime[row.name], axis=1)

multiplier = 20
while len(df) < 1000:
    p=random.randint(2,100)
    toMultiply=df[p-1:p]
    toMultiply = pd.concat([toMultiply] * multiplier, ignore_index = True)
    arTime = random_dates('2024-04-01', '2024-04-30', multiplier, 's', seed=None)
    toMultiply["date_of_sale"]=toMultiply.apply(lambda row: arTime[row.name], axis=1)

    df = pd.concat([df, toMultiply], ignore_index = True)


df = df.sample(frac=1, ignore_index= True)
# df.reset_index(drop=True)

# df.index += 1
df[:1000].to_csv('../src/main/resources/gameTitlesOneK.csv',index=False)
```
